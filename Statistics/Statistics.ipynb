Statistics Questions:-

 1-What is hypothesis testing in statistics?
  -> Hypothesis testing in statistics is a method used to make inferences or draw conclusions about a population based on sample data. The goal      is to test whether there is enough statistical evidence to support or reject a hypothesis about the population.

2- What is the null hypothesis, and how does it differ from the alternative hypothesis?
  -> Null Hypothesis (H₀):
     The null hypothesis is a statement of no effect, no difference, or no relationship. It's the default position that assumes nothing unusual      is happening in the population or data being studied.It represents the idea that any observed effect or difference is due to random            chance or sampling variability.The null hypothesis is often framed to suggest that a certain parameter (such as a mean, proportion, or         difference) is equal to a specific value.In hypothesis testing, the null hypothesis is typically what we aim to test against or fail to         reject.
     Alternative Hypothesis (H₁ or Ha):
     The alternative hypothesis is a statement that suggests there is an effect, difference, or relationship. It’s what you want to prove or         find evidence for.This hypothesis posits that the population parameter is different from what’s stated in the null hypothesis.
     If the null hypothesis is rejected, the alternative hypothesis is supported.

3- What is the significance level in hypothesis testing, and why is it important?
   ->The significance level (denoted as α) is a critical concept in hypothesis testing. It defines the threshold for determining whether the       results of a statistical test are statistically significant or not. In other words, it helps you decide how much evidence is needed to        reject the null hypothesis.The significance level (α) is a probability value that represents the likelihood of rejecting the null             hypothesis when it is actually true (also known as a Type I error).It’s a threshold you set before conducting the test, usually between       0 and 1, with the most common value being 0.05 (5%).If α = 0.05, this means you are willing to accept a 5% chance of wrongly rejecting        the null hypothesis when it is true.
     The significance level is important for several reasons:
     1-Setting a Criteria for Evidence: It helps you decide how strong the evidence needs to be before you reject the null hypothesis. If the        p-value (the probability of observing the data, or something more extreme, under the null hypothesis) is less than or equal to the            significance level, you reject the null hypothesis.
     2-Control of Type I Error:By setting α, you're essentially controlling the rate at which you might make a Type I error (false positive).        For example, if you set α = 0.05, you’re saying that you are willing to accept a 5% chance of concluding the alternative hypothesis is        true when, in fact, the null hypothesis is true.
     3-Interpreting Statistical Results:The significance level helps you interpret the p-value:
       If the p-value ≤ α, you reject the null hypothesis (indicating sufficient evidence for the alternative hypothesis).If the p-value > α,        you fail to reject the null hypothesis (indicating insufficient evidence to support the alternative hypothesis).
     4-Comparing Different Tests or Studies:The significance level allows for consistency and comparability between different tests or               studies, providing a common baseline for determining statistical significance.Example:
       Let’s say you’re testing whether a new drug has an effect on blood pressure (the null hypothesis being that the drug has no effect).If        you set α = 0.05, you are accepting a 5% risk of concluding the drug has an effect when, in reality, it doesn’t (i.e., rejecting the          null hypothesis when it's actually true).If your p-value from the test is 0.03, which is less than 0.05, you would reject the null            hypothesis and conclude there is evidence that the drug has an effect.If your p-value is 0.07, which is greater than 0.05, you would          fail to reject the null hypothesis and conclude there isn’t enough evidence to support that the drug has an effect.
     5-Balancing the Significance Level:Lower α values (e.g., 0.01) decrease the risk of Type I errors but make it harder to reject the null         hypothesis, meaning you may miss detecting a true effect (Type II error).Higher α values (e.g., 0.10) increase the risk of Type I             errors but make it easier to reject the null hypothesis, potentially detecting effects that aren’t actually there.

   

4- What does a P-value represent in hypothesis testing?
   -> The p-value is a key concept in hypothesis testing. It helps you determine the strength of the evidence against the null hypothesis.
      A p-value is the probability of obtaining a test statistic (like a sample mean, difference between groups, etc.) at least as extreme as       the one actually observed, assuming the null hypothesis is true. In other words, it tells you how likely your data (or something more         extreme) would occur if there were actually no effect, difference, or relationship in the population (i.e., if the null hypothesis were       true).

5- How do you interpret the P-value in hypothesis testing?
   -> Interpreting the p-value in hypothesis testing is crucial for drawing meaningful conclusions from statistical data.
      1. P-value ≤ α (Significance Level):
        Interpretation: Reject the null hypothesis.
        What it means: The observed data is sufficiently unlikely under the null hypothesis. There’s enough evidence to conclude that the            alternative hypothesis is more likely to be true.
      2. P-value > α (Significance Level)
        Interpretation: Fail to reject the null hypothesis.
        What it means: The observed data is likely under the null hypothesis. There isn’t strong enough evidence to conclude that the                 alternative hypothesis is true. It does not mean the null hypothesis is proven true, just that there’s insufficient evidence against          it.

6-What are Type 1 and Type 2 errors in hypothesis testing?
  ->In hypothesis testing, Type I and Type II errors are two types of mistakes that can occur when making conclusions about the null              hypothesis. They are related to incorrect decisions about rejecting or failing to reject the null hypothesis based on the sample data.
    1. Type I Error (False Positive):
        Definition: A Type I error occurs when you reject the null hypothesis when it is actually true.
        What it means: You incorrectly conclude that there is an effect or a difference when, in reality, there isn’t. It's a false positive          because you’re detecting something that doesn’t exist.
        Probability: The probability of committing a Type I error is denoted by the significance level (α). For example, if α = 0.05, there           is a 5% chance of making a Type I error.
    2. Type II Error (False Negative)
        Definition: A Type II error occurs when you fail to reject the null hypothesis when it is actually false.
        What it means: You incorrectly conclude that there is no effect or difference when, in reality, there is one. It’s a false negative           because you fail to detect something that is actually there.
        Probability: The probability of committing a Type II error is denoted by β (beta). The power of a test, which is 1 - β, represents            the probability of correctly rejecting the null hypothesis when it is false.

7- What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   -> In hypothesis testing, the difference between a one-tailed test and a two-tailed test lies in the direction of the hypothesis you are         testing.
    1. One-Tailed Test:
       A one-tailed test tests for the possibility of the relationship in one direction (either greater than or less than).The alternative           hypothesis (H₁) specifies the direction of the effect. For example:Right-tailed test: Tests if the parameter is greater than a certain        value. Example: H₁: μ > μ₀ (the population mean is greater than a specific value).
       Left-tailed test: Tests if the parameter is less than a certain value. Example: H₁: μ < μ₀ (the population mean is less than a                specific value).
       A one-tailed test only considers evidence in one direction. If the data falls in the opposite direction, it will not lead to rejecting        the null hypothesis (H₀).It is used when the researcher has a specific expectation about the direction of the effect.
    2. Two-Tailed Test:
       A two-tailed test tests for the possibility of the relationship in both directions (either greater than or less than).The alternative         hypothesis (H₁) is non-directional. Example: H₁: μ ≠ μ₀ (the population mean is not equal to a specific value).A two-tailed test looks        at both ends of the distribution to see if the test statistic is too extreme in either direction.It is used when the researcher does          not have a specific expectation of the direction of the effect and wants to test for any significant difference, regardless of whether        it’s an increase or decrease.
       Example:One-tailed: You believe that a new drug will increase the average recovery rate, so your hypothesis might be that the recovery        rate with the new drug is greater than with the current treatment.
        H₀: μ ≤ μ₀ (the drug does not improve recovery rate).
        H₁: μ > μ₀ (the drug improves recovery rate).
        Two-tailed: You just want to know if the new drug has a different effect, whether it increases or decreases the recovery rate.    
        H₀: μ = μ₀ (the drug has no effect).
        H₁: μ ≠ μ₀ (the drug has an effect, either positive or negative).

8- What is the Z-test, and when is it used in hypothesis testing?
   -> A Z-test is a type of statistical test used in hypothesis testing when the population variance is known (or the sample size is large          enough for the Central Limit Theorem to apply), and the data follows a normal distribution. It compares the sample data to the                population parameter (such as the mean or proportion) to determine if there is a significant difference between them.
      A Z-test is used in the following situations:
      1-When comparing a sample mean to a population mean (for a one-sample Z-test).
        Example: You want to test if the average height of a sample of people is different from the known average height of the population.
      2-When comparing two sample means (for a two-sample Z-test), assuming the population variances are known or large sample sizes.  
        Example: You want to test if the average test scores of two groups of students are significantly different.
      3-When comparing sample proportions to population proportions (for a Z-test for proportions).
        Example: You want to test if the proportion of people who prefer a certain product is different from the proportion in the general            population.

It is used in hypothesis testing to standardize test statistics, helping to determine      how extreme or unusual a particular sample result is, relative to the null hypothesis.
     The formula for calculating the Z-score depends on the type of test you're conducting, but it is generally:
                  Z= (X−μ)/σ
            Where:
               𝑋= the value of the sample statistic (e.g., sample mean, observed value)
               𝜇= the population mean (or hypothesized value in hypothesis testing)
               𝜎= the population standard deviation (or standard error if using sample data)
      the Z-Score Represent:-
      1-The Z-score tells you how many standard deviations the sample statistic (like the sample mean or sample proportion) is away from the          population mean under the null hypothesis.
      2-A positive Z-score indicates the sample statistic is above the population mean.
      3-A negative Z-score indicates the sample statistic is below the population mean.

10- What is the T-distribution, and when should it be used instead of the normal distribution?
   ->The T-distribution (also known as Student’s t-distribution) is a probability distribution that is similar to the normal distribution          but with heavier tails. It is used in hypothesis testing and confidence interval estimation when the sample size is small and/or the          population standard deviation is unknown.
     You should use the T-distribution instead of the normal distribution in the following situations:
     1-Small Sample Sizes: When the sample size is small (𝑛<30), the sample mean and sample standard deviation are more likely to deviate           from the true population values. The T-distribution accounts for this uncertainty.
     2-Unknown Population Standard Deviation: If you do not know the population standard deviation (𝜎), you must use the sample standard            deviation (𝑠) as an estimate. This introduces additional uncertainty, and the T-distribution is used to adjust for this.
     3-When the Data is Normally Distributed: Like the normal distribution, the T-distribution assumes that the data is approximately                normally distributed. If the data is heavily skewed or contains outliers, neither the normal nor T-distribution might be appropriate,         and other non-parametric tests may be needed.

12- What is the T-test, and how is it used in hypothesis testing?
   -> A T-test is a statistical test used to determine if there is a significant difference between the means of two groups or a sample mean        and a population mean, especially when the population standard deviation is unknown and the sample size is small (typically 𝑛<30).
      It is called the T-test because it uses the T-distribution (also known as Student’s t-distribution) to determine whether the difference       between the sample statistic and the population parameter is statistically significant.
      The general steps to conduct a T-test are as follows:
      1-State the Hypotheses:
        Null Hypothesis (H₀): The null hypothesis assumes that there is no significant difference between the sample mean and the population                                mean (or between the means of two groups). It represents the status quo.    
        Alternative Hypothesis (H₁): The alternative hypothesis states that there is a significant difference between the sample mean and the         population mean (or between the two sample means).
      2-Choose a Significance Level (α):
        This is typically set to 0.05 (5%), which indicates that you’re willing to accept a 5% chance of incorrectly rejecting the null               hypothesis (Type I error).
      3- Collect Data and Calculate the T-Statistic:
         The T-statistic depends on the type of T-test you're performing.
      4- Find the Critical Value:
         Use a T-distribution table or statistical software to find the critical value corresponding to the degrees of freedom and the                 significance level (𝛼).
      5- Make the Decision:
        Compare the absolute value of your calculated t-statistic to the critical value.If the t-statistic is greater than the critical               value, reject the null hypothesis.If the t-statistic is less than or equal to the critical value, fail to reject the null  hypothesis.
      6- Interpret the Results:
        If the null hypothesis is rejected, there is significant evidence to support the alternative hypothesis.

14- What is a confidence interval, and how is it used to interpret statistical results?
    -> A confidence interval (CI) is a range of values, derived from sample data, that is used to estimate a population parameter (like a            population mean or proportion). It provides a range of plausible values for the population parameter, with an associated level of             confidence that the true value lies within this range.For example, if you estimate the mean height of students in a school and compute        a 95% confidence interval, this means you're 95% confident that the true population mean height lies within the interval.
       A confidence interval provides more information than a simple point estimate (like a sample mean). It helps in assessing the precision        and uncertainty of an estimate. Here's how it is interpreted:
       1.Estimation of Population Parameter:
         The confidence interval gives a range of values within which the true population parameter (e.g., population mean, proportion) is             likely to fall.
       2.Confidence Level:
         The confidence level (e.g., 95%) indicates the likelihood that the interval contains the true population parameter. A 95% confidence          interval means that if you were to take 100 random samples and compute a confidence interval for each one, about 95 of those                  intervals would contain the true population parameter.
       3.Precision of the Estimate:
         Wider intervals indicate more uncertainty about the true population parameter (i.e., lower precision), while narrower intervals               suggest more precision.A larger sample size reduces the margin of error, leading to a narrower confidence interval (higher                    precision). Conversely, a smaller sample size leads to a wider interval.
       4.Statistical Significance:
          A confidence interval can be used to determine statistical significance. If a confidence interval for the difference between two              means (for example, in a two-sample t-test) does not contain zero, it suggests that there is a statistically significant difference           between the groups.
       

15-What is the margin of error, and how does it affect the confidence interval?
   ->The margin of error (MoE) is a measure of the uncertainty or potential error in a statistical estimate. It indicates how much the sample      statistic (such as a sample mean or proportion) is expected to vary from the true population parameter due to sampling variability. The       margin of error is essentially the "range" that is added and subtracted from the sample estimate to create the confidence interval.
     The margin of error plays a crucial role in determining the width of the confidence interval. Here's how it affects the CI:
     1.Wider Margin of Error = Wider Confidence Interval:
        A larger margin of error means that the interval is wider, indicating more uncertainty or variability in the estimate. This generally         happens when:
        The confidence level is higher (e.g., 99% instead of 95%).
        The sample size is smaller.
        The standard deviation of the population (or sample) is large.
     2.Smaller Margin of Error = Narrower Confidence Interval:
        A smaller margin of error results in a narrower confidence interval, which suggests that the estimate is more precise and there is            less uncertainty about the population parameter. This occurs when:
        The confidence level is lower (e.g., 90% instead of 95%).
        The sample size is larger.
        The standard deviation of the population (or sample) is smaller. 

16-How is Bayes' Theorem used in statistics, and what is its significance?
   -> Bayes' Theorem is a fundamental concept in statistics and probability theory that provides a way to update the probability of a               hypothesis based on new evidence. It combines prior knowledge (or beliefs) about a situation with new data to refine our understanding        of that situation. In simple terms, it helps us revise our predictions when we get more information.
       The Formula:    Bayes' Theorem is written as:
                               𝑃(𝐻∣𝐸)=(𝑃(𝐸∣𝐻)⋅𝑃(𝐻))/𝑃(𝐸)
        Where:
              𝑃(𝐻∣𝐸) is the posterior probability, which is the probability of the hypothesis 𝐻 being true given the evidence 𝐸.
              𝑃(𝐸∣𝐻) is the likelihood, or the probability of observing the evidence 𝐸 given that the hypothesis 𝐻 is true.
              𝑃(𝐻) is the prior probability, or the initial probability of the hypothesis 𝐻 before seeing the evidence.
              𝑃(𝐸) is the marginal likelihood, or the total probability of the evidence under all possible hypotheses.
        Significance in Statistics:
        1-Incorporating Prior Knowledge: Bayes' Theorem allows statisticians to incorporate prior beliefs or data into their analysis. This             is especially helpful when dealing with limited or uncertain data. The prior can be based on historical data, expert opinion, or              any other reasonable assumption.  
        2-Updating with New Data: As new data (evidence) becomes available, Bayes' Theorem provides a way to update probabilities. This makes           it highly useful in situations where the situation is dynamic, and data is continuously evolving.
        3-Inference and Decision Making: In statistics, Bayes' Theorem is used for drawing inferences and making decisions under uncertainty.           By continuously updating our beliefs about the likelihood of different outcomes, we can make more informed decisions. For example,            in medical diagnostics, it can help update the probability that a patient has a particular disease based on test results. 
        4-Machine Learning: In machine learning, especially in Bayesian models, the theorem is used in algorithms for classification,                   prediction, and model selection. It forms the basis for many probabilistic models, such as Naive Bayes classifiers.
        5-Uncertainty Quantification: The theorem helps to quantify uncertainty. It doesn't just give a single "best guess" for a probability           but provides a way to consider all possible outcomes and how they change with new evidence.

20- What is an ANOVA test, and what are its assumptions?
    -> An Analysis of Variance (ANOVA) test is a statistical method used to determine if there are statistically significant differences             between the means of three or more groups. It is used when you want to test whether different treatments, conditions, or categories           produce different outcomes. ANOVA compares the variance within each group to the variance between groups to assess whether the group          means differ more than would be expected by chance.
       To ensure the validity of the results from an ANOVA test, several assumptions must be met:
       1-Independence: The observations within each group must be independent of each other. This means that one observation in a group                              should not influence or be related to another observation. 
       2-Normality: The data within each group should be approximately normally distributed. ANOVA is robust to moderate deviations from                          normality, but significant deviations can affect the results, particularly with small sample sizes.       
       3-Homogeneity of Variance (Homoscedasticity): The variance within each group should be roughly equal across all groups. This                                 assumption ensures that the groups are comparable in terms of the variability in their data. This is typically tested                          with Levene's test or Bartlett's test.        
       4-Random Sampling: The data should come from a random sample, meaning the groups should be randomly selected from the population.

21- What are the different types of ANOVA tests?
   -> There are several types of ANOVA (Analysis of Variance) tests, each designed for different experimental designs and research questions.       The choice of ANOVA test depends on the number of factors (independent variables) involved and whether the factors are manipulated or         observed.the main types of ANOVA tests:
      1- One-Way ANOVA (Single-Factor ANOVA): Used when you are comparing the means of three or more independent groups based on one factor                                                 or independent variable.
      2-Two-Way ANOVA (Factorial ANOVA): Used when you have two independent variables (factors) and you want to examine their individual                                               effects on a dependent variable as well as any potential interaction between the two factors.
      3-Repeated Measures ANOVA: Used when the same subjects are measured multiple times or under different conditions. This type of ANOVA                                     accounts for the correlation between repeated measurements on the same subject.
      4-Mixed-Design ANOVA (Split-Plot ANOVA): A combination of Two-Way ANOVA and Repeated Measures ANOVA. It’s used when there are both                                                    within-subjects and between-subjects factors in the study. This is useful when you have both                                                  repeated measures on some factors and different groups for other factors.
      5-Multivariate Analysis of Variance (MANOVA): Used when there are multiple dependent variables and you want to assess the effects of                                                       one or more independent variables on all of these dependent variables simultaneously.                                                         MANOVA is an extension of ANOVA for multivariate data.

22- What is the F-test, and how does it relate to hypothesis testing?
    ->The F-test is a statistical test that is used to compare two variances and assess whether the variances of two populations are                significantly different from each other. It is also used in the context of ANOVA (Analysis of Variance) to test if there are                  significant differences between group means.
      the F-test Works in Hypothesis Testing:
      1- Step 1: State the Hypotheses:
         For an ANOVA F-test, you start with the null hypothesis that all group means are equal and the alternative hypothesis that at least           one group mean differs.For a two-sample F-test, the null hypothesis is that the variances of the two populations are equal, and the           alternative hypothesis is that they are different.
      2- Step 2: Calculate the F-statistic:      
         For ANOVA, the F-statistic is calculated by dividing the variance between groups by the variance within groups. This is the ratio of          the variance due to the treatment or group differences (between-group variance) to the variance within the groups (error or residual          variance).For a two-sample F-test, the F-statistic is the ratio of the two sample variances.
      3- Step 3: Find the Critical Value:
         Using the degrees of freedom from both the numerator and denominator, you can look up the critical value of F from an F-distribution          table at a chosen significance level (typically 0.05).
      4- Step 4: Compare the F-statistic to the Critical Value:
         If the calculated F-statistic is greater than the critical value, you reject the null hypothesis.
         If the F-statistic is smaller than the critical value, you fail to reject the null hypothesis.
      5- Step 5: Conclusion:

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

# Function to calculate and visualize Type 1 and Type 2 errors
def visualize_errors(population_mean_null, population_mean_alternate, population_std, sample_size, alpha=0.05):
    # Step 1: Calculate critical value (for two-tailed test)
    critical_value = stats.norm.ppf(1 - alpha / 2)

    # Step 2: Calculate the Z-scores for Type 1 and Type 2 errors
    # For Type 1 Error: Under the null hypothesis, we reject the null hypothesis if the sample mean falls beyond the critical value.
    # For Type 2 Error: Under the alternative hypothesis, we fail to reject the null hypothesis if the sample mean falls inside the acceptance region.

    # Step 3: Define the normal distributions for the null and alternative hypothesis
    null_distribution = stats.norm(loc=population_mean_null, scale=population_std / math.sqrt(sample_size))
    alt_distribution = stats.norm(loc=population_mean_alternate, scale=population_std / math.sqrt(sample_size))

    # Step 4: Calculate the Type 1 Error (area under the null distribution that falls in the rejection region)
    type_1_error_area_left = null_distribution.cdf(-critical_value)  # Left rejection region
    type_1_error_area_right = 1 - null_distribution.cdf(critical_value)  # Right rejection region
    type_1_error = type_1_error_area_left + type_1_error_area_right

    # Step 5: Calculate the Type 2 Error (area under the alternative distribution that falls in the acceptance region)
    acceptance_region_left = critical_value
    acceptance_region_right = -critical_value
    type_2_error_area = alt_distribution.cdf(acceptance_region_left) - alt_distribution.cdf(acceptance_region_right)
    
    # Step 6: Print the results
    print(f"Type 1 Error: {type_1_error:.4f}")
    print(f"Type 2 Error: {type_2_error_area:.4f}")

    # Step 7: Visualization
    x = np.linspace(population_mean_null - 4 * population_std, population_mean_alternate + 4 * population_std, 1000)

    # Plot the null and alternative distributions
    plt.plot(x, null_distribution.pdf(x), label="Null Hypothesis Distribution", color='blue')
    plt.plot(x, alt_distribution.pdf(x), label="Alternative Hypothesis Distribution", color='green')

    # Shade the rejection regions for Type 1 error
    plt.fill_between(x, null_distribution.pdf(x), where=(x <= -critical_value), color='red', alpha=0.5, label="Rejection Region (Left, Type 1 Error)")
    plt.fill_between(x, null_distribution.pdf(x), where=(x >= critical_value), color='red', alpha=0.5, label="Rejection Region (Right, Type 1 Error)")

    # Shade the acceptance region for Type 2 error
    plt.fill_between(x, alt_distribution.pdf(x), where=(x >= -critical_value) & (x <= critical_value), color='yellow', alpha=0.5, label="Acceptance Region (Type 2 Error)")

    # Add labels and legend
    plt.axvline(x=-critical_value, color='black', linestyle='dashed', label=f"Critical Value: ±{critical_value:.2f}")
    plt.axvline(x=critical_value, color='black', linestyle='dashed')
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
population_mean_null = 100  # Population mean under the null hypothesis
population_mean_alternate = 105  # Population mean under the alternative hypothesis
population_std = 15  # Population standard deviation
sample_size = 30  # Sample size

# Visualize Type 1 and Type 2 errors
visualize_errors(population_mean_null, population_mean_alternate, population_std, sample_size)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
import scipy.stats as stats

# Function to perform an Independent T-test
def independent_t_test(group1, group2, alpha=0.05):
    # Step 1: Calculate the sample means, variances, and sample sizes
    mean1 = np.mean(group1)
    mean2 = np.mean(group2)
    var1 = np.var(group1, ddof=1)  # Sample variance (ddof=1 for unbiased estimator)
    var2 = np.var(group2, ddof=1)
    n1 = len(group1)
    n2 = len(group2)
    
    # Step 2: Calculate the pooled standard deviation (if variances are assumed equal)
    pooled_std = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    
    # Step 3: Calculate the t-statistic
    t_stat = (mean1 - mean2) / (pooled_std * np.sqrt(1 / n1 + 1 / n2))
    
    # Step 4: Calculate the degrees of freedom
    df = n1 + n2 - 2
    
    # Step 5: Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))
    
    # Step 6: Print results
    print(f"Group 1 Mean: {mean1}, Group 2 Mean: {mean2}")
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

    # Step 7: Interpret the results
    if p_value < alpha:
        print(f"Reject the null hypothesis: There is a significant difference between the two groups.")
    else:
        print(f"Fail to reject the null hypothesis: There is no significant difference between the two groups.")

# Simulate two groups of data
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)  # Group 1: mean=50, std=10, size=30
group2 = np.random.normal(loc=55, scale=12, size=30)  # Group 2: mean=55, std=12, size=30

# Perform the Independent T-test
independent_t_test(group1, group2, alpha=0.05)
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence_level=0.95):
    """
    Calculate the confidence interval for the sample mean.
    
    Parameters:
    - sample (array-like): The sample data.
    - confidence_level (float): The confidence level (default is 0.95).
    
    Returns:
    - tuple: Lower bound and upper bound of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    sample_size = len(sample)
    
    # Degrees of freedom for the t-distribution
    df = sample_size - 1
    
    # t-critical value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df)
    
    # Margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
    
    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Sample data (mean=50, std=10, n=30)

# Calculate 95% confidence interval for the sample mean
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval for the sample mean: ({lower:.2f}, {upper:.2f})")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, likelihood, false_positive_rate, evidence):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: The prior probability of the hypothesis (probability the patient has the disease).
    - likelihood: The likelihood of the evidence given the hypothesis (probability of test being positive given the patient has the disease).
    - false_positive_rate: The probability of test being positive given the patient does not have the disease.
    - evidence: The overall probability of the test being positive (which combines both true positives and false positives).
    
    Returns:
    - posterior: The posterior probability of the hypothesis given the evidence.
    """
    # Calculate the posterior using Bayes' Theorem
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.01  # Prior probability: 1% of the population has the disease
likelihood = 0.9  # Likelihood: 90% chance the test is positive given the disease
false_positive_rate = 0.05  # False positive rate: 5% chance the test is positive without the disease

# Total probability of the test being positive (evidence) is calculated as:
# P(D) = P(Test Positive | Disease) * P(Disease) + P(Test Positive | No Disease) * P(No Disease)
evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))

# Calculate the posterior probability using Bayes' Theorem
posterior = bayes_theorem(prior, likelihood, false_positive_rate, evidence)

# Output the result
print(f"The posterior probability that the patient has the disease given a positive test result is: {posterior:.4f}")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test from the observed data.
    
    Parameters:
    - observed_data (2D array or DataFrame): The observed frequency table (contingency table).
    
    Returns:
    - expected_frequencies (2D array): The expected frequency table based on the assumption of independence.
    """
    # Convert observed data to a numpy array (if it's a DataFrame, it's automatically converted)
    observed = np.array(observed_data)
    
    # Get the row totals and column totals
    row_totals = observed.sum(axis=1)  # Sum of rows
    column_totals = observed.sum(axis=0)  # Sum of columns
    
    # Total number of observations
    total = observed.sum()
    
    # Calculate the expected frequencies using the formula
    expected_frequencies = np.outer(row_totals, column_totals) / total
    
    return expected_frequencies
# Example data: A 2x2 contingency table
# Assume we're testing if gender (Male/Female) is independent of the preference for a type of drink (Coffee/Tea)
data = {'Coffee': [50, 30], 'Tea': [30, 60]}
observed_table = pd.DataFrame(data, index=['Male', 'Female'])

# Calculate expected frequencies
expected_table = calculate_expected_frequencies(observed_table)

# Display the observed and expected frequencies
print("Observed Frequencies:")
print(observed_table)
print("\nExpected Frequencies:")
print(expected_table)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
import scipy.stats as stats

def chi_square_goodness_of_fit(observed, expected):
    """
    Perform a Chi-square goodness-of-fit test.

    Parameters:
    - observed: The observed frequencies (list or array).
    - expected: The expected frequencies (list or array).
    
    Returns:
    - chi2_stat: The Chi-square statistic.
    - p_value: The p-value for the test.
    """
    # Adjust expected frequencies to sum to the same total as observed frequencies
    total_observed = observed.sum()
    total_expected = expected.sum()
    
    # Adjust the expected frequencies proportionally
    adjusted_expected = expected * (total_observed / total_expected)
    
    # Calculate the Chi-square statistic and p-value
    chi2_stat, p_value = stats.chisquare(observed, adjusted_expected)
    
    return chi2_stat, p_value

# Example data: Observed and expected frequencies
observed = np.array([30, 50, 20])  # Observed data (Vanilla, Chocolate, Strawberry)
expected = np.array([33.33, 33.33, 33.33])  # Expected distribution (equal preference)

# Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chi_square_goodness_of_fit(observed, expected)

# Output the results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data fits the expected distribution.")
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1: The first sample (list or array).
    - sample2: The second sample (list or array).
    - alpha: Significance level (default is 0.05).

    Returns:
    - f_statistic: The calculated F-statistic.
    - p_value: The p-value for the F-test.
    - decision: The decision to reject or fail to reject the null hypothesis.
    """
    # Calculate the sample variances
    var1 = np.var(sample1, ddof=1)  # Sample variance of sample1
    var2 = np.var(sample2, ddof=1)  # Sample variance of sample2
    
    # F-statistic: larger variance / smaller variance
    if var1 > var2:
        f_statistic = var1 / var2
    else:
        f_statistic = var2 / var1
    
    # Degrees of freedom for the numerator and denominator
    df1 = len(sample1) - 1  # Degrees of freedom for sample1
    df2 = len(sample2) - 1  # Degrees of freedom for sample2
    
    # Calculate the p-value using the F-distribution
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))
    
    # Make the decision
    if p_value < alpha:
        decision = "Reject the null hypothesis: The variances are significantly different."
    else:
        decision = "Fail to reject the null hypothesis: The variances are equal."
    
    return f_statistic, p_value, decision

np.random.seed(42)

# Generate two random samples
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1 with mean=50, std=10
sample2 = np.random.normal(loc=50, scale=15, size=100)  # Sample 2 with mean=50, std=15

# Perform the F-test
f_statistic, p_value, decision = f_test(sample1, sample2)

# Output the results
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(decision)
```

```python
#17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.
import numpy as np
import scipy.stats as stats

def anova_test(*groups):
    """
    Perform a one-way ANOVA test to compare the means between multiple groups.
    
    Parameters:
    - *groups: The groups of data to compare (passed as arguments).
    
    Returns:
    - f_statistic: The F-statistic for the ANOVA test.
    - p_value: The p-value for the ANOVA test.
    - decision: The decision to reject or fail to reject the null hypothesis.
    """
    # Perform the ANOVA test
    f_statistic, p_value = stats.f_oneway(*groups)
    
    # Decision based on p-value
    alpha = 0.05  # Significance level
    if p_value < alpha:
        decision = "Reject the null hypothesis: At least one group mean is different."
    else:
        decision = "Fail to reject the null hypothesis: All group means are equal."
    
    return f_statistic, p_value, decision

# Example usage:
np.random.seed(42)

# Simulate data for 3 groups (e.g., three different treatment groups)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=12, size=30)  # Mean=55, Std=12
group3 = np.random.normal(loc=60, scale=15, size=30)  # Mean=60, Std=15

# Perform the ANOVA test
f_statistic, p_value, decision = anova_test(group1, group2, group3)

# Output the results
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(decision)
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check assumptions for One-Way ANOVA
def check_anova_assumptions(*groups):
    """
    Checks the assumptions for One-Way ANOVA:
    - Normality: Shapiro-Wilk test
    - Homogeneity of variances: Levene's test
    - Independence: This is assumed to be satisfied based on study design
    
    Parameters:
    - *groups: The groups of data to check assumptions for.
    
    Returns:
    - normality_results: List of results for Shapiro-Wilk test (Normality)
    - homogeneity_results: Result of Levene's test (Homogeneity of Variances)
    """
    normality_results = []
    
    # Check normality using Shapiro-Wilk test
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        normality_results.append((f"Group {i+1}", stat, p_value))
        
    # Check homogeneity of variances using Levene's test
    stat, p_value_levene = stats.levene(*groups)
    # Interpret results
    normality_decisions = [f"{group[0]} - Normality (Shapiro-Wilk) p-value: {group[2]:.4f}" for group in normality_results]
    normality_decision = "Normality assumption passed for all groups." if all(group[2] > 0.05 for group in normality_results) else "Normality assumption violated for at least one group."

    homogeneity_decision = "Homogeneity of variances assumption passed." if p_value_levene > 0.05 else "Homogeneity of variances assumption violated."

    return normality_results, normality_decision, homogeneity_decision

# Example usage:
np.random.seed(42)

# Simulate data for 3 groups
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=12, size=30)  # Mean=55, Std=12
group3 = np.random.normal(loc=60, scale=15, size=30)  # Mean=60, Std=15

# Check assumptions for One-Way ANOVA
normality_results, normality_decision, homogeneity_decision = check_anova_assumptions(group1, group2, group3)

# Print normality results
for result in normality_results:
    print(f"{result[0]}: Shapiro-Wilk Statistic = {result[1]:.4f}, p-value = {result[2]:.4f}")

# Print decisions
print(normality_decision)
print(homogeneity_decision)

# Visualization of Normality using Q-Q plots
plt.figure(figsize=(12, 6))
# Q-Q plot for each group
for i, group in enumerate([group1, group2, group3], 1):
    plt.subplot(1, 3, i)
    stats.probplot(group, dist="norm", plot=plt)
    plt.title(f"Q-Q Plot for Group {i}")

plt.tight_layout()
plt.show()
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
import numpy as np
from scipy import stats

# Step 1: Simulate data
np.random.seed(42)

# Simulate a sample from a normal distribution
sample_size = 50
sample_data = np.random.normal(loc=50, scale=10, size=sample_size)  # Mean = 50, SD = 10

# Calculate the sample variance
sample_variance = np.var(sample_data, ddof=1)

# Hypothesized population variance
population_variance = 100

# Step 2: Perform the Chi-square test for variance
# Chi-square statistic formula: (n - 1) * sample_variance / population_variance
chi_square_stat = (sample_size - 1) * sample_variance / population_variance

# Degrees of freedom: n - 1
df = sample_size - 1

# Calculate the p-value for a two-tailed test
# We need to find the probability that the Chi-square statistic is greater than the observed statistic
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Step 3: Output results
print(f"Sample Variance: {sample_variance}")
print(f"Chi-square Statistic: {chi_square_stat}")
print(f"P-value: {p_value}")

# Step 4: Interpretation
alpha = 0.05  # Significance level

if p_value < alpha:
    print("We reject the null hypothesis: The variance is significantly different from the hypothesized variance.")
else:
    print("We fail to reject the null hypothesis: The variance is not significantly different from the hypothesized variance.")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
import numpy as np
from scipy import stats

# Step 1: Sample Data
# For example, we have data from two groups
# Group 1: 100 trials, 60 successes
# Group 2: 120 trials, 72 successes

n1 = 100  # Sample size for group 1
x1 = 60   # Successes in group 1

n2 = 120  # Sample size for group 2
x2 = 72   # Successes in group 2

# Step 2: Calculate sample proportions
p1_hat = x1 / n1
p2_hat = x2 / n2

# Step 3: Calculate the pooled proportion
p_pool = (x1 + x2) / (n1 + n2)

# Step 4: Calculate the Z-test statistic
z_stat = (p1_hat - p2_hat) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

# Step 5: Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

# Step 6: Output results
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Step 7: Interpretation
alpha = 0.05  # Significance level

if p_value < alpha:
    print("We reject the null hypothesis: The proportions are significantly different.")
else:
    print("We fail to reject the null hypothesis: The proportions are not significantly different.")
```
